#3. Opening a csv file from an URL

Load file from any server

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)
```

#11. Skip bad lines

Resolving error like :- Expected 8 fields in line 6452, saw 9


Pandas replaced:

error_bad_lines,
warn_bad_lines

with a single cleaner parameter → on_bad_lines

Options you can use

'error' → (default) raises error

'warn' → skips bad lines + shows warning

'skip' → silently skips bad lines

#14. Convertors

```python
#I want short form of team name like royal challenges bangloare --> rcb
pd.read_csv('/content/IPL Matches 2008-2020.csv')
```

```python
def rename(name):
  if name == "Royal Challengers Bangalore":
    return "RCB"
  else:
    return name
```

```python
pd.read_csv('/content/IPL Matches 2008-2020.csv',converters={'team1':rename})
```
