```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Flatten
```

```python
(x_train,y_train),(x_test,y_test) = keras.datasets.mnist.load_data()
```

```python
import matplotlib.pyplot as plt
plt.imshow(x_train[0])
```

```python
#so the all values comes in range of 0 to 1
x_train = x_train/255
x_test = x_test/255
```

```python
x_train[0]
```

In this we have data like 28 * 28 but for model training we have to flatten the data in 1D 784

```python
model = Sequential()

model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128,activation = 'relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer = 'Adam')
```

```python
model.fit(x_train,y_train,epochs=10,validation_split=0.2)
```

```python
y_prob = model.predict(x_test)
```

```python
y_prob
```

In Above result we have 10 layers in our neural network which gives us the probability of having that image 0 or 1 .... or 9. so we have to predict which no. is in image so that we can to see max value of probability in each 1d array.

```python
y_pred = y_prob.argmax(axis=1)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
model1 = Sequential()

model1.add(Flatten(input_shape=(28,28)))
model1.add(Dense(128,activation = 'relu'))
model1.add(Dense(32,activation = 'relu'))
model1.add(Dense(10,activation='softmax'))
```

```python
model1.summary()
```

```python
model1.compile(loss='sparse_categorical_crossentropy',optimizer = 'Adam',metrics=['accuracy'])
```

```python
history = model1.fit(x_train,y_train,epochs=25,validation_split=0.2)
```

```python
y_prob = model1.predict(x_test)
```

```python
y_pred = y_prob.argmax(axis=1)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```
